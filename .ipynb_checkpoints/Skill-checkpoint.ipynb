```python
class LinearRegressionGD(object) :
    def __init__(self , eta=0.001 , n_iter=20) :
        self.eta = eta
        self.n_iter = n_iter
        
    def fit(self, X, у):
        self.w_ = np.zeros(1 + X.shape[1])
        self.cost_ = []

        for i in range(self.n_iter):
            output = self.net_input(X)
            errors = (у - output)
            self.w_[1:] += self.eta * X.T.dot(errors)
            self.w_[0] += self.eta * errors.sum()
            cost = (errors**2).sum() / 2.0
            self.cost_.append(cost)
        return self

    def net_input(self, Х):
        return np.dot(X , self.w_[1:]) + self.w_[0]
    
    def predict(self, Х) :
        return self.net_input(X)
```

```python
df["LeagueIndex"] = df["LeagueIndex"].apply(lambda x: df["APM"].mode() if x == '?' else x)
df["APM"] = df["APM"].apply(lambda x: df["APM"].mean() if x == '?' else x)
df["SelectByHotkeys"] = df["SelectByHotkeys"].apply(lambda x: df["SelectByHotkeys"].mean() if x == '?' else x)
df["NumberOfPACs"] = df["NumberOfPACs"].apply(lambda x: df["NumberOfPACs"].mean() if x == '?' else x)
df["GapBetweenPACs"] = df["GapBetweenPACs"].apply(lambda x: df["GapBetweenPACs"].mean() if x == '?' else x)
df["ActionLatency"] = df["ActionLatency"].apply(lambda x: df["ActionLatency"].mean() if x == '?' else x)

# for k, v in df:
#     df[k] = df[k].apply(lambda x: df[k].mean() if x == '?' else x)
```

```python
X=df[['APM']].values
y=df[['SelectByHotkeys']].values
from sklearn.linear_model import LinearRegression
slr = LinearRegression()
slr.fit(X,y)
print('Наклон: %.3f' % slr.coef_[0])
print('Перечение: %.3f' % slr.intercept_)
```

```python
def lin_regplot (X, у, model) :
    plt.scatter(X, у , c= 'blue' )
    plt.plot(X, model.predict(X), color= 'red')
    return None
lin_regplot(X, y, slr)
plt.xlabel('[APM]')
plt.ylabel('[SelectByHotkeys] ')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
sc_y = StandardScaler()
X_std = sc_x.fit_transform(X)
y_std = sc_y.fit_transform(y)

slr.fit(X_std,y_std)
lin_regplot(X_std, y_std, slr)
plt.xlabel('[APM]')
plt.ylabel('[SelectByHotkeys] ')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
Х = df.iloc[ :, :-1 ].values
у = df['APM'].values
X_std = sc_x.fit_transform(X)
y_std = sc_y.fit_transform(y)
X_train , X_test, y_train , y_test = train_test_split(
X_std, y_std , test_size=0.3 , random_state=0)
slr = LinearRegression()
slr.fit(X_train , y_train)
y_train_pred = slr.predict(X_train)
y_test_pred = slr.predict(X_test)

from sklearn.metrics import mean_squared_error
print('MSE тренировка: %.3f, тестирование: %.3f' % (
mean_squared_error(y_train, y_train_pred),
mean_squared_error(y_test, y_test_pred)))
```

```python
from sklearn.metrics import r2_score
print('R2 тренировка: %.3f, тестирование: %.3f' %(
r2_score(y_train, y_train_pred),
r2_score(y_test, y_test_pred)))
```

```python
# ЧТО ПРОИСХОДИТ?
X=df[['APM']].values
y=df[['SelectByHotkeys']].values
from sklearn.preprocessing import PolynomialFeatures
quadratic = PolynomialFeatures(degree=2)
regr = LinearRegression()
cubic = PolynomialFeatures(degree=3)


X_quad = quadratic.fit_transform(X)
X_cubic = cubic.fit_transform(X)

# linear fit
X_fit = np.arange(X.min(), X.max(), 1)[:, np.newaxis]
regr = regr.fit(X, y)
y_lin_fit = regr.predict(X_fit)
linear_r2 = r2_score(y, regr.predict(X))
# quadratic fit
regr = regr.fit(X_quad, y)
y_quad_fit = regr.predict(quadratic.fit_transform(X_fit))
quadratic_r2 = r2_score(y, regr.predict(X_quad))
# cubic fit
regr = regr.fit(X_cubic, y)
y_cubic_fit = regr.predict(cubic.fit_transform(X_fit))
cubic_r2 = r2_score(y, regr.predict(X_cubic))
# plot results
plt.scatter(X, y,
            label='training points',
            color='lightgray')
plt.plot(X_fit, y_lin_fit,
            label='linear (d=1), $R^2=%.2f$'
            % linear_r2,
            color='blue',
            lw=2,
            linestyle=':')
plt.plot(X_fit, y_quad_fit,
            label='quadratic (d=2), $R^2=%.2f$'
            % quadratic_r2,
            color='red',
            lw=2,
            linestyle='-')
plt.plot(X_fit, y_cubic_fit,
            label='cubic (d=3), $R^2=%.2f$'
            % cubic_r2,
            color='green',
            lw=2,
            linestyle='--')
plt.xlabel('APM')
plt.ylabel('SelectByHotkeys')
plt.legend(loc='upper left')
plt.show()












# X_quad = quadratic.fit_transform(X_train)

# # квадратичная подгонка
# X_fit= np.arange(X.min(), X.max(), 1) [: , np.newaxis]
# lr = lr.fit(X_quad, y_train)
# y_quad_fit = lr.predict(quadratic.fit_transform(X_fit))
# quadratic_r2 = r2_score(y ,lr.predict(X_quad))
# y_quad_pred = lr.predict(X_quad)

# print('train MSE %.3f ' % 
# mean_squared_error(y, y_quad_pred))
# print('train R2 : %.3f' % r2_score(y, y_quad_pred))
```

```python
X = df.iloc[:, :-1].values
y = df['APM'].values
X_train, X_test, y_train, y_test =train_test_split(X, y,test_size=0.3, random_state=0)
from sklearn.ensemble import RandomForestRegressor
forest = RandomForestRegressor(n_estimators=1000,criterion='mse',random_state=0, n_jobs=-1)
forest.fit(X_train, y_train)
y_train_pred = forest.predict(X_train)
y_test_pred = forest.predict(X_test)
print('MSE тренировка: %.3f, тестирование: %.3f' % (
mean_squared_error(y_train, y_train_pred),
mean_squared_error(y_test, y_test_pred)))
print('R2 тренировка: %.3f, тестирование: %.3f' %(
r2_score(y_train, y_train_pred),
r2_score(y_test, y_test_pred)))

print("Важность признаков:")
feat_labels=df.columns[1:]
importances = forest.feature_importances_
indices = np.argsort(importances)[::-1]
for f in range(X_train.shape[1]):
    print ("%2d) %-*s %f" % (f + 1, 25 ,feat_labels[indices[f]] ,importances[indices[f]]))
```
